```python
import numpy as np
import pandas as pd
```

```python
living_pop_2019_12 = pd.read_csv('dec.csv')
```

```python
living_pop_2019_12.columns
```

```python
living_pop_2019_12 = living_pop_2019_12.groupby(['행정동코드', '시간대구분'],as_index=False).mean()
```

```python
living_pop_2019_12.columns.tolist()[4:18]
```

```python
male = living_pop_2019_12.columns.tolist()[4:18]

female = living_pop_2019_12.columns.tolist()[18:]
```

```python
living_pop_2019_12['male'] = living_pop_2019_12[male].sum(axis=1)
living_pop_2019_12['female'] = living_pop_2019_12[female].sum(axis=1)
```

```python
living_pop_2019_12 = living_pop_2019_12.rename(columns={'행정동코드':'adm_code',
                                                        '시간대구분':'time',
                                                        '총생활인구수':'total'})
```

```python
living_pop_2019_12 = living_pop_2019_12[['adm_code','time','total','male','female']]
```

```python
living_pop_2019_12_pivot = living_pop_2019_12.pivot(index='adm_code', columns='time',values=['total','male','female'])
```

```python
living_pop_2019_12_pivot.columns.levels[1]
```

```python
columns = []

for p in living_pop_2019_12_pivot.columns.levels[0]:
    for i in living_pop_2019_12_pivot.columns.levels[1]:
        columns.append(p+'_'+str(i))
```

```python
living_pop_2019_12_pivot.columns = columns
```

```python
living_pop_2019_12_pivot['total_avg']=living_pop_2019_12_pivot[living_pop_2019_12_pivot.columns[:24]].mean(axis=1)
```

```python
living_pop_2019_12_pivot['male_avg']=living_pop_2019_12_pivot[living_pop_2019_12_pivot.columns[24:48]].mean(axis=1)
```

```python
living_pop_2019_12_pivot['female_avg']=living_pop_2019_12_pivot[living_pop_2019_12_pivot.columns[48:72]].mean(axis=1)
```

```python
living_pop_2019_12_pivot['gu'] = np.array(living_pop_2019_12_pivot.index)//1000
```

```python
living_pop_2019_12_pivot
```

```python
coord = pd.read_csv('dong_coord.csv',index_col=0)
```

```python
coord = coord[['adm_cd2','lat','lng']]
```

```python
coord.adm_cd2 = coord.adm_cd2//100
```

```python
living_pop_2019_12_pivot = living_pop_2019_12_pivot.merge(coord,left_index=True, right_on='adm_cd2')
```

```python
living_pop_2019_12_pivot = living_pop_2019_12_pivot.set_index('adm_cd2')
```
