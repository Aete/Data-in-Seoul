# 1. Subway

### 1) Split dataset.

<br />
In this part, I split dataset to several CSV files by monthly. In addition to this, I also made a CSV file about an average of years.

As a result of this part, the CSV files are supposed to save at 'csvfiles' folder. The CSV files were in sub-folders by years 2017-2019.

**If you already have the files which is equal to files in the  'csvfiles' folder folder, you don't need to do this part**

Dataset: The average number of passengers on the subway by station and time (monthly) <br />
Link : [The average number of passengers on the subway by station and time](https://data.seoul.go.kr/dataList/datasetView.do?infId=OA-12252&srvType=S&serviceKind=1&currentPageNo=1)

```python
# import Libraries
import pandas as pd
import numpy as np
```

```python
# read csv file
passenger_monthly = pd.read_csv('passenger_num(subway).csv',encoding = 'utf8')
passenger_monthly
```

```python
# this part is to chage column names to english

passenger_monthly = passenger_monthly.drop('작업일자',axis=1)
passenger_monthly = passenger_monthly.rename(columns={'사용월' : 'month',
                                                      '호선명' : 'line',
                                                      '지하철역' : 'station'
})
new_col_dict = {}
for i, col in enumerate(passenger_monthly.columns[3:]):
    new_col = col[:2]+'-'+col[4:6]
    
    if i % 2 == 0:
        new_col = new_col + ' (on)'
    else:
        new_col = new_col + ' (off)'
    
    new_col_dict[col] = new_col

passenger_monthly = passenger_monthly.rename(columns=new_col_dict)
```

```python
# this part is to split 'month' column to 'year' and 'month'

passenger_monthly['year']= passenger_monthly['month']//100
passenger_monthly['month']= passenger_monthly['month']%100
passenger_monthly['month'] = passenger_monthly['month'].astype('int')
columns = passenger_monthly.columns
passenger_monthly = pd.concat([passenger_monthly[columns[-1]],passenger_monthly[columns[:-1]]],axis=1)
```

```python
passenger_monthly.head()
```

```python
# import Libraries
station_location = pd.read_csv('Station_location.csv',sep=',')
```

```python
# removing columns which we don't need

target_to_remove_list = ['전철역코드','호선','외부코드','사이버스테이션','X좌표','Y좌표']
station_location = station_location.drop(target_to_remove_list,axis=1)

# removing columns which we don't need
station_location  = station_location.rename(columns={'전철역명' : 'station',
                                                     'X좌표(WGS)' : 'X(WGS)',
                                                     'Y좌표(WGS)' : 'Y(WGS)'
                                                     
})
```

```python
# this dict is about stations which don't have any location information.

missing_station = {}
missing_station['419민주묘지'] = [37.649532, 127.013658]
missing_station['여주'] = [37.282419, 127.628537]
missing_station['가오리'] = [37.641971, 127.016770]
missing_station['북한산보국문'] = [37.612288, 127.008222]
missing_station['북한산우이'] = [37.663155, 127.012767]
missing_station['삼양'] = [37.626962, 127.018148]
missing_station['삼양사거리'] = [37.621334, 127.020489]
missing_station['솔밭공원'] = [37.656098, 127.013227]
missing_station['솔샘'] = [37.620259, 127.013567]
missing_station['정릉'] = [37.602659, 127.013477]
missing_station['화계'] = [37.634111, 127.017493]
missing_station['지평'] = [37.476453, 127.629649]

# put the information at station_location dataframe
for s in missing_station.keys():
    station_location.loc[station_location['station']==s,'X(WGS)'] = missing_station[s][0]
    station_location.loc[station_location['station']==s, 'Y(WGS)']= missing_station[s][1]
    
```

```python
def data_spliter(df, year):
    '''
    This function is to split dataframe to CSV files by months and average of the year.
    As a result of this function, we will get CSV files like below:
    1. CSV file about the number of subway passenger by monthly.
     - Passenger information is organized by time as a given csv file, but I add monthly average passenger number.
    2. CSV file about an average number of subway passenger in the specific year.
     - Passenger information is organized by time as a given csv file, and I also add monthly average passenger number.
    
    Args:
    df : target dataframe
    year : num of year which you want to change

    Returns: None
    '''
    dataset = df[df['year']==year]
    month_list = dataset['month'].unique().tolist()
    column_list = dataset.columns.tolist()
    on_list = [columns[s] for s in range(4,len(columns),2)]
    off_list = [columns[s] for s in range(5,len(columns),2)]
    dataset['Total(on)'] = dataset[on_list].sum(axis=1)
    dataset['Total(off)'] = dataset[off_list].sum(axis=1)
    dataset = pd.merge(dataset,station_location,left_on='station',right_on='station')
    output_address1 = 'csvfiles/'+str(year)+'/'
    for month in month_list:
        dataset_monthly = dataset[dataset['month']==month]
        output_address2 = output_address1 + str(month)+'.csv'
        dataset_monthly.to_csv(output_address2,sep=',',index=False)
    average_by_years = dataset.groupby(['line','station'],as_index=False).mean().drop(['month'],axis=1)
    average_by_years.to_csv(output_address1+'average.csv',sep=',',index=False)
```

```python
import os
for year in range(2015,2020):
    if os.path.isdir('csvfiles/'+str(year)) == False:
        os.mkdir('csvfiles/'+str(year))
    data_spliter(passenger_monthly, year)
```

### 2) Data Visualization (folium).

<br />
This part is to visualize data with folium library.

As a result of this part, html files are supposed to save at 'html(folium)' folder. The html files were in sub-folders by years 2017-2019.

To get location of each station, I will use several dataset which contain station id and location.

Dataset: The subway station id and location <br />
Link : [The subway station id and location](https://data.seoul.go.kr/dataList/datasetView.do?infId=OA-118&srvType=S&serviceKind=1&currentPageNo=1)

```python
import pandas as pd

passenger_num_2019 = pd.read_csv('csvfiles/2019/average.csv',sep=',')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(pd.DataFrame(passenger_num_2019['total(on)']))
```

```python
import pandas as pd
import folium
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import os
target_year_list = [2017,2018,2019]
for year in target_year_list:
    input_file = 'csvfiles/'+str(year)+'/average.csv'
    output_address = 'html(folium)/'+str(year)+'/average.html'
    base_map = folium.Map(location=[37.565822, 126.976880], tiles="Stamen Toner", zoom_start=15)
    passenger_num = pd.read_csv(input_file,sep=',')
    passenger_num['total(on)_scaled'] = scaler.transform(pd.DataFrame(passenger_num['total(on)']))
    station_list = passenger_num['station'].unique().tolist()

    for station in station_list:
        target_ = passenger_num[passenger_num['station']== station]
        X_ = target_['X(WGS)'].tolist()[0]
        Y_ = target_['Y(WGS)'].tolist()[0]
        station_name = station
        line_list = target_['line'].tolist()
        popup_text = '<b>'+station+'</b>' + '&nbsp monthly passenger(on)<br>'
        for line in line_list:
            popup_text += '<span>{} : {}</span><br>'.format(line,int(target_[target_['line']==line]['total(on)']))
        folium.Circle(
                      location=[X_,Y_],
                      radius= np.log1p(target_['total(on)_scaled'].sum())*500,
                      popup = folium.Popup(popup_text,max_width=300),
                      color='crimson',
                      fill=True,
                      fill_color='crimson'
                    ).add_to(base_map)

    if os.path.isdir('html(folium)/'+str(year)) == False:
        os.mkdir('html(folium)/'+str(year))
    base_map.save(output_address)
```

```python
import plotly
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

plotly.tools.set_credentials_file(username='##', api_key='##)
mapbox_access_token = '##'
passenger_num_2019 = pd.read_csv('csvfiles/2019/average.csv',sep=',')
scaler = MinMaxScaler()
scaler.fit(pd.DataFrame(passenger_num_2019['total(on)']))
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.grid_objs import Grid, Column

station_list = passenger_num_2019['station'].unique().tolist()
lon_list = []
lat_list = []
size_list_on = []
size_list_off = []
for i in range(len(on_list)):
    on_size = on_size_list[i]
    off_size = off_size_list[i]
    passenger_num_2019[on_size] = scaler.transform(pd.DataFrame(passenger_num_2019[on_list[i]]))
    passenger_num_2019[off_size] = scaler.transform(pd.DataFrame(passenger_num_2019[off_list[i]]))

for station in station_list:
    target_ = passenger_num_2019[passenger_num_2019['station']==station]
    lat_list.append(target_['X(WGS)'].tolist()[0])
    lon_list.append(target_[passenger_num_2019['station']==station]['Y(WGS)'].tolist()[0])
    size_list_on.append((target_[on_size_list].sum(axis=0)*500).tolist())
    size_list_off.append((target_[off_size_list].sum(axis=0)*500).tolist())

graph_on = go.Scattermapbox(lat = lat_list,
                            lon = lon_list,
                            mode = 'markers',
                            text = station_list,  
                            hoverinfo = 'text',
                            marker = dict(size = size_list_on[5],
                                          opacity = 0.7,
                                          color = '#2196F3')
                            )

graph_off = go.Scattermapbox(lat = lat_list,
                              lon = lon_list,
                              mode = 'markers',
                              text = station_list, 
                              marker = dict(size = size_list_off[5],
                                            opacity = 0.7,
                                            color = '#F44336')
                            )

data = [graph_on,graph_off]
```

```python
updatemenus = list([
    dict(type="buttons",
         active = 0,
         buttons=list([
            dict(label ='Both',
                 method='update',
                 args = [{'visible':[True,True]},
                         {'title' : 'Passenger Number (On&Off)'}]),
            dict(label ='On',
                 method='update',
                 args = [{'visible':[True,False]},
                         {'title' : 'Passenger Number (On)'}]),
            dict(label ='Off',
                 method='update',
                 args = [{'visible':[False,True]},
                         {'title' : 'Passenger Number (Off)'}])
         ]),
        direction = 'down',
        pad = {'r': 10, 't': 10}, 
        x = 0.01,
        xanchor = 'left',
        y = 1,
        yanchor = 'top',
        bgcolor = '#AAAAAA',
        bordercolor = '#FFFFFF',
        font = dict(size=11)
        )
])
```

```python
sliders = dict(active = 0,
               yanchor = 'bottom',
               xanchor = 'right',
               currentvalue = {
                                'font': {'size': 20},
                                'prefix': 'time',
                                'visible': True,
                                'xanchor': 'right'
                            },   
                transition={'duration': 300, 'easing': 'cubic-in-out'},
                pad= {'b': 10, 't': 50},
                len= 0.9,
                x= 0.1,
                y= 0,
                steps= [])


```

```python
layout = go.Layout(
    title = 'Passenger Number of Subway Station (On&Off)',
    autosize=True,
    showlegend=False,
    hovermode='closest',
    updatemenus=updatemenus,
    sliders=[sliders],
    mapbox=go.layout.Mapbox(
        accesstoken=mapbox_access_token,
        style = 'dark',
        bearing=0,
        center=go.layout.mapbox.Center(

            lat=37.565822,
            lon=126.976880
        ),
        pitch=0,
        zoom=13
    ),
)
```

```python
fig = go.Figure(data=data, layout=layout)
py.plot(fig,filename='sampleplot.html')
```
