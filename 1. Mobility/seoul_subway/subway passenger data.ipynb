### 1. Subway

### 1) covert daily passenger data to monthly

In this part, I convert daily passenger data to monthly

Dataset: The average number of passengers on the subway by station and time (monthly) <br />
Link : [The average number of passengers on the subway by station and time](https://data.seoul.go.kr/dataList/datasetView.do?infId=OA-12252&srvType=S&serviceKind=1&currentPageNo=1)

```python
# import Libraries
import pandas as pd
import numpy as np
```

```python
# read csv file
passenger_monthly = pd.read_csv('passenger_num(subway).csv',encoding = 'utf8')
passenger_monthly
```

```python
passenger_monthly = passenger_monthly.drop('작업일자',axis=1)
passenger_monthly = passenger_monthly.rename(columns={'사용월' : 'month',
                                                      '호선명' : 'line',
                                                      '지하철역' : 'station'
})
```

```python
new_col_dict = {}
for i, col in enumerate(passenger_monthly.columns[3:]):
    new_col = col[:2]+'-'+col[4:6]
    
    if i % 2 == 0:
        new_col = new_col + ' (on)'
    else:
        new_col = new_col + ' (off)'
    
    new_col_dict[col] = new_col

passenger_monthly = passenger_monthly.rename(columns=new_col_dict)
```

```python
passenger_monthly 
```

```python
passenger_monthly['year']= passenger_monthly['month']//100
passenger_monthly['month']= passenger_monthly['month']%100
passenger_monthly['month'] = passenger_monthly['month'].astype('int')
```

```python
columns = passenger_monthly.columns
```

```python
passenger_monthly
```

```python
passenger_monthly = pd.concat([passenger_monthly[columns[-1]],passenger_monthly[columns[:-1]]],axis=1)
```
